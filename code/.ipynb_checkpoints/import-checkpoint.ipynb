# Recommendation of books
Feed your favorite books then I will give you recommended books that you would like

Theorem of this app

* Get rating of each book
* Find review of your favorite book
* Find books that the reviewer who left good rate on your favorite book evaluate as good book

I will use `cosine distance` to see how similar a book is to others

---

## Import dataset

* book - book title, book id
* user - user id
* rate - user id, book id and rating

```python
import pandas as pd
```

```python
book = pd.read_csv('../data/BX_Books.csv', sep=';', encoding='latin-1')
book.head()
```

```python
user = pd.read_csv('../data/BX-Users.csv', sep=';', encoding='latin-1')
user.head()
```

```python
rate = pd.read_csv('../data/BX-Book-Ratings.csv', sep=';', encoding='latin-1')
rate.head()
```

---

## Clean and EDA

### Merge all datasets

```python
df = pd.merge(user, rate, how='inner', left_on='User-ID', right_on='User-ID')
df = pd.merge(df, book, how='inner', left_on='ISBN', right_on='ISBN')
df.head()
```

## Assemble dataframe

```python
# FInd missing values
df.isnull().sum()
```

```python
df.shape
```

There are tons of rows and this is too big to assemble my app. I will used only infuential rows which has

* No missing values
* User-ID who rates at least 10 books

This app refers reviews which are given by same user so if a user left only one review, it is not influential for my app

```python
# Drop all missing values
df.dropna(inplace=True)
```

```python
# Get index where a user left at least 10 retings
reviewer_index = df['User-ID'].value_counts().index[df['User-ID'].value_counts()>=10]
```

```python
# New dataframe with above index
mini = df.iloc[reviewer_index]
mini.reset_index(inplace=True, drop=True)
mini.head()
```

```python
mini.shape
```

### Create pivot table to see relation of books, users and ratings

```python
pivot = mini.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pivot.head()
```

### Generate `cosine distance`

```python
from scipy import sparse
pivot_sparse = sparse.csr_matrix(pivot.fillna(0))
```

```python
from sklearn.metrics.pairwise import pairwise_distances, cosine_distances, cosine_similarity
cos_distance = pairwise_distances(pivot_sparse, metric='cosine')
recommender = pd.DataFrame(cos_distance, columns=pivot.index, index=pivot.index)
recommender.head()
```
